```python

import matplotlib.pyplot as plt
```

```python
car=np.array(['farari','Bugaty','BMW','Tesla','maruti'])
```

```python
cost=np.array([12,24,14,44,74])
year=np.array([2001,2002,2003,2001,2005])
```

```python
plt.plot(car,cost,'go--')
plt.title("Shap")
plt.grid = (True)
plt.ylabel('year')
plt.xlabel('Size')
plt.show()
```

```python
plt.bar(car,cost,)
```

```python
plt.subplot(1,2,1)

plt.plot(car,cost,'go--')
plt.title("Shap")
plt.ylabel('year')
plt.xlabel('Size')

plt.subplot(1,2,2)
plt.bar(car,cost)
```

```python
plt.subplot(1,1,1)

plt.plot(car,cost,'go--')
plt.title("Shap")
plt.ylabel('year')
plt.xlabel('Size')

plt.subplot(1,1,1)
plt.bar(car,cost)
```

```python

explode =(0.4,0.6,0.8,0.2,0)
plt.pie(cost, explode=explode,labels=car, autopct='%1.1f%%',shadow=True, startangle=87)

plt.show()
```
